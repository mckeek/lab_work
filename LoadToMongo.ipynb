```python
#get the data from the NBA.com/stats 

import json
import urllib.request
import requests
nbastats = "http://stats.nba.com/stats/leaguedashplayerstats?College=&Conference=&Country=&DateFrom=&DateTo=&Division=&DraftPick=&DraftYear=&GameScope=&GameSegment=&Height=&LastNGames=0&LeagueID=00&Location=&MeasureType=Base&Month=0&OpponentTeamID=0&Outcome=&PORound=0&PaceAdjust=N&PerMode=PerGame&Period=0&PlayerExperience=&PlayerPosition=&PlusMinus=N&Rank=N&Season=2018-19&SeasonSegment=&SeasonType=Regular+Season&ShotClockRange=&StarterBench=&TeamID=0&VsConference=&VsDivision=&Weight="
r = requests.get(nbastats,headers={'User-Agent':'Mozilla/5.0'})
data = r.json()
#data = json.load(r)

#print(json.dumps(data))
for d in data:
    print(d)
```

```python
#find the Orlando Magic Players so we can make sure they are there in the DB

PDB =  {}
ORL = {}
ORLs = {}
player = data['resultSets']
for p in player:
    plist = p['rowSet']
    for item in plist:
        PID = item[0]
        Pname = item[1]
        Tname = item[3]
        if Tname == "ORL":
            ORL[Pname] = PID
            ORLs[Pname] = item
        PDB[Pname] = Tname
#print(PDB)
       
#print(ORL)

print(ORL)
```

```python
#we cant use load so we have to combine the columns and data into a df and then back to a dict
#this is the data and the columns
pdata = data['resultSets'][0]['rowSet']
columnsx = data['resultSets'][0]['headers']

#print(pdata)
print(pdata)

#for p in pdata:
#   print(p)
```

```python
#create the df using pdata and columnsx
import pandas as pd
from pandas import DataFrame as df
import numpy as np
d = []
for p in pdata:
    d.append(p)
playerdf = pd.DataFrame(d, columns=(columnsx))
cols = ['AGE', 'GP', 'W', 'L', 'W_PCT', 'MIN', 'FGM', 'FGA', 'FG_PCT', 'FG3M', 'FG3A', 'FG3_PCT', 'FTM', 'FTA', 'FT_PCT', 'OREB', 'DREB', 'REB', 'AST', 'TOV', 'STL', 'BLK', 'BLKA', 'PF', 'PFD', 'PTS', 'PLUS_MINUS', 'NBA_FANTASY_PTS', 'DD2', 'TD3', 'GP_RANK', 'W_RANK', 'L_RANK', 'W_PCT_RANK', 'MIN_RANK', 'FGM_RANK', 'FGA_RANK', 'FG_PCT_RANK', 'FG3M_RANK', 'FG3A_RANK', 'FG3_PCT_RANK', 'FTM_RANK', 'FTA_RANK', 'FT_PCT_RANK', 'OREB_RANK', 'DREB_RANK', 'REB_RANK', 'AST_RANK', 'TOV_RANK', 'STL_RANK', 'BLK_RANK', 'BLKA_RANK', 'PF_RANK', 'PFD_RANK', 'PTS_RANK', 'PLUS_MINUS_RANK', 'NBA_FANTASY_PTS_RANK', 'DD2_RANK', 'TD3_RANK', 'CFID', 'CFPARAMS']


#print(playerdf)
playerdf[cols] = playerdf[cols].apply(pd.to_numeric, errors='coerce', axis=1)

#print(playerdf)




playerdf.loc[playerdf['TEAM_ABBREVIATION'] == 'ORL']
```
