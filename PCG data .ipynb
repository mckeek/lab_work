```python
import pandas as pd
y2018 = pd.read_csv('fadata/2018-Table 1.csv')
y2017 = pd.read_csv('fadata/2017-Table 1.csv')
y2016 = pd.read_csv('fadata/2016-Table 1.csv')
y2015 = pd.read_csv('fadata/2015-Table 1.csv')
y2014 = pd.read_csv('fadata/2014-Table 1.csv')
```

```python
frames = [y2014,y2015,y2016,y2017,y2018]
result = pd.concat(frames)
```

```python

result['price'] = result['Unnamed: 1'].str.replace('\t', '')
result = result.iloc[1:]
result = result.drop(['Unnamed: 1', 'Unnamed: 2', 'Unnamed: 3','Unnamed: 4'], axis=1)
result.columns = ['date', 'gas-price']
result.sort_values('date')
#result

```
